# Yfinance sandbox

```python
import yfinance as yf   
from forex_python.converter import CurrencyRates

import datetime
```

## Get currency data

```python
cr = CurrencyRates()
sgd_to_usd = cr.get_rate('SGD', 'USD', datetime.datetime(2026, 1, 2))
sgd_to_usd
```

```python
# Using yfinance to get exchange data instead
fx = yf.download("SGDUSD=X", start=datetime.datetime(2026, 1, 1), end=datetime.datetime(2026, 1, 10))  # End-Date exclusive

# Use Close as FX rate
fx = fx[['Close']].rename(columns={'Close': 'FXRate'})
print(fx.stack(level=1).reset_index())
```
